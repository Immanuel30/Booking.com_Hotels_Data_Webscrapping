```python
#Importing libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
import psycopg2
```

```python
#Base Url for webscrapping
base_url = "https://www.booking.com/"
```

```python
#Testing the response from the base URL

response= requests.get(base_url)

if response.status_code ==200:
    print("Request was successfull")
    soup = BeautifulSoup(response.text, 'html.parser') 
else:
    print("Request failed")
```

```python
soup
```

```python
# Extracting all the URLs in the page
all_urls= []
soup.find_all('a')

for link in soup.find_all('a'):
    all_urls.append(link.get('href'))
```

```python
all_pics = []

soup.find_all('a276c777b0')

for pics in soup.find_all('a276c777b0'):
    all_pics.append(pics.get('scr'))
```

```python
all_urls
```

```python
#Extracting all the country URLs
country_urls = [url for  url in all_urls if 'country' in url]
```

```python
country_urls
```

```python
def get_data(country_urls):
    response = requests.get(country_urls)
    if response.status_code ==200:
        print("Request was successfull")
        soup = BeautifulSoup(response.text, 'html.parser') 
    else:
        print("Request failed")
    return soup
```

```python
ng_urls =  []

for link in country_urls.find_all('a'):
    if '/ng' in link.get('href'):
        ng_urls.append(link.get('href'))
```
